```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
#now providing a summary of sales distribution with the use statistical analysis tools
```

```python
sales_summary = {
    "mean": np.mean(df['Sales']),
    "median": np.median(df['Sales']),
    "std": np.std(df['Sales']),
    "min": np.min(df['Sales']),
    "max": np.max(df['Sales'])
}
print("Sales Summary:\n", sales_summary)
```

```python
#now proceeding towards profit distribution :
```

```python
profit_summary = {
    "mean": np.mean(df['Profit']),
    "median": np.median(df['Profit']),
    "std": np.std(df['Profit']),
    "min": np.min(df['Profit']),
    "max": np.max(df['Profit'])
}
print("Profit Summary:\n", profit_summary)
```

```python
#now we will determine sales through different measures:
```

```python
sales_by_category = df.groupby('Category')['Sales'].sum()
print("Sales by Category:\n", sales_by_category)
```

```python
sales_by_sub_category = df.groupby('Sub-Category')['Sales'].sum()
print("Sales by Sub-Category:\n", sales_by_sub_category)
```

```python
sales_by_region = df.groupby('Region')['Sales'].sum()
print("Sales by Region:\n", sales_by_region)
```

```python
df_grouped = df.groupby('Order Date')['Sales'].sum().reset_index()
print("Sales Trends Over Time:\n", df_grouped.head())
```

```python
top_cities = df.groupby('City')['Sales'].sum().nlargest(10)
print("Top 10 Cities by Sales:\n", top_cities)
```

```python
high_value_transactions = df[df['Sales'] > np.percentile(df['Sales'], 95)]
print("High-Value Transactions (Top 5%):\n", high_value_transactions)
```

```python
aov = np.mean(df['Sales'] / df['Quantity'])
print("Average Order Value (AOV):", aov)
```

```python
df['Year-Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Year-Month')['Sales'].sum()
print("Monthly Sales Trends:\n", monthly_sales)
```

```python
#now proving some visuals using matplot:
```

```python
category_sales = df.groupby('Category')['Sales'].sum()
plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales.values, color='maroon')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
top_cities = df.groupby('City')['Sales'].sum().nlargest(10)
plt.figure(figsize=(10,5))
plt.bar(top_cities.index, top_cities.values, color='orange')
plt.title("Top 10 Cities by Sales")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
df['Year-Month'] = df['Order Date'].dt.to_period('M')
monthly_sales = df.groupby('Year-Month')['Sales'].sum()
plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index.astype(str), monthly_sales.values, linestyle='solid', marker='o', color='brown')
plt.title("Monthly Sales Trends")
plt.xlabel("Year-Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(8,5))
plt.bar(region_sales.index, region_sales.values, color='lightgreen')
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', colors=['red', 'green', 'blue', 'lightgreen'])
plt.title("Sales Distribution by Region")
plt.show()
```

```python
df_region_trends = df.groupby(['Order Date', 'Region'])['Sales'].sum().unstack()
plt.figure(figsize=(12,6))
for region in df_region_trends.columns:
    plt.plot(df_region_trends.index, df_region_trends[region], linestyle='solid', marker='o', label=region)
plt.title("Sales Trends Over Time by Region")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
country_orders = df.groupby('Country')['Order ID'].count().sort_values(ascending=False)
plt.figure(figsize=(10,5))
plt.bar(country_orders.index, country_orders.values, color='turquoise')
plt.title("Orders by Country")
plt.xlabel("Country")
plt.ylabel("Total Orders")
plt.xticks(rotation=45)
plt.show()
```
